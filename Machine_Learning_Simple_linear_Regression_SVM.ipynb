```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Import Dataset**

```
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1:].values
```

## **Spliting dataset**


```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```
y_test
```

## **Simple Linear regression on Training set**

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

## **Predicting on Test set**

```
y_pred= regressor.predict(x_test)
```

```
y_test
```

## **Visualization of Training set**

```
plt.scatter(x_train,y_train, color = "red")
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## **Visulization of Test set**

```
plt.scatter(x_test,y_test, color = "Blue")
plt.plot(x_train, regressor.predict(x_train), color="Red")
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```
 from sklearn.svm import SVR
```

```
regresor_svm = SVR(kernel="linear")
```

```
regresor_svm.fit(x,y)
```

```
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y), color = 'red')
plt.plot(sc_x.inverse_transform(x), sc_y.inverse_transform(regresor_svm.predict(x)), color ='blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```
sc_y.inverse_transform(regresor_svm.predict(sc_x.transform([[10.5]])))
```
